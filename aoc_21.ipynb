```python
#3 
def power_consumption(vlist):

    #sum cols
    lista = []
    for col in range(len(vlist[0])):
        basket = 0
        for row in range(len(vlist)):
            basket = (ord(vlist[row][col]) - ord("0")) + basket
        lista.append(basket)

    #1 or 0 the most common? check sum and compare to n/2; if <n/2 then 0, else: 1
    epsilon = []
    for val in lista:
        if val > len(vlist)/2:
            epsilon.append(1)
        else:
            epsilon.append(0)

    #now define gama epsilon==0 to 1 and epsilon==1 to 0
    gamma = []
    for val in epsilon:
        gamma.append(abs(val-1))

    #gamma list to str
    strings = [str(integer) for integer in gamma]
    gamma_string = "". join(strings)

    #epsilon list to str
    strings = [str(integer) for integer in epsilon]
    epsilon_string = "". join(strings)

    #epsilon & gamma from binary to int & return 
    return int(epsilon_string, 2)*int(gamma_string,2)

vlist = ["00100",
"11110",
"10110",
"10111",
"10101",
"01111",
"00111",
"11100",
"10000",
"11001",
"00010",
"01010"]


power_consumption(vlist)
```

```python
#7
lista = [16,1,2,0,4,2,7,1,2,14]

#selection sort algorithm (for practice purposes...)
def minimo(lista, start):
    minimo = lista[start]
    pos = start
    for val in range(start+1, len(lista)):
        if lista[val] < minimo:
            minimo = lista[val]
            pos = val
    return (minimo, pos)

def sort(lista):
    for val in range(len(lista)):
        temp = lista[val]
        pair = minimo(lista, val)
        lista[val] = pair[0]
        lista[pair[1]] = temp
    return lista

#since we now the max; now we cal calculate sum of the diff of each element to the max range

def distance(lista):
    lista_dist = []
    for ele in range(sort(lista)[-1]+1):
        ele_ = []
        for val in range(len(lista)):
            ele_.append(abs(sort(lista)[val]-ele))
        lista_dist.append(ele_)
    return lista_dist

#now we have a matrix with the vectors; we only have to sum rows and retrieve the index lowest one

def soma(lista):
    if lista == []:
        return 0
    return lista[0] + soma(lista[1:])

# retrieve the minimum sum
def row_sum(lista):
    lista = distance(lista)
    minimo = soma((lista)[0])
    index = 0
    for row in range(len(lista)):
        if soma(lista[row]) < minimo:
            minimo = soma(lista[row])
            index = row

    return minimo, index

row_sum(lista)
    
```
